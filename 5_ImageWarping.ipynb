
3. Write a function of the form (10%):



>> imwarped = warpImage(im,H,limit_coordinates)



where im is the input image to be warped, H is the homography, and limit_coordinates defines the limits (xmin,xmax,ymin,ymax) of the warped image.  


```python
import numpy as np
import cv2

def warpImage(im, H, limit_coordinates):

    xmin, xmax, ymin, ymax = limit_coordinates
    height, width, channels = im.shape

    # Create coordinate grid
    x = np.arange(width)
    y = np.arange(height)
    xv, yv = np.meshgrid(x, y)

    # Convert to homogeneous coordinates
    grid = np.vstack([xv.ravel(), yv.ravel(), np.ones_like(xv.ravel())])

    # Apply homography transformation
    newgrid = H @ grid
    newgrid /= newgrid[2, :]  # Convert from homogeneous to Cartesian

    # Round coordinates and convert to int
    x_new = newgrid[0, :].round().astype(int)
    y_new = newgrid[1, :].round().astype(int)

    # Define output image with given limits
    width_out = xmax - xmin + 1
    height_out = ymax - ymin + 1
    imwarped = np.full((height_out, width_out, channels), -1, dtype=np.int16)

    # Valid pixel selection
    valid_x = np.logical_and(x_new >= xmin, x_new < xmax)
    valid_y = np.logical_and(y_new >= ymin, y_new < ymax)
    valid = np.logical_and(valid_x, valid_y)

    # Map pixels from original image to warped image
    imwarped[y_new[valid] - ymin, x_new[valid] - xmin] = im[yv.ravel()[valid], xv.ravel()[valid]]

    # Convert -1 values to valid colors using inpainting
    mask = (imwarped[:, :, 0] == -1).astype(np.uint8) * 255
    imwarped = np.clip(imwarped, 0, 255).astype(np.uint8)
    imwarped = cv2.inpaint(imwarped, mask, inpaintRadius=3, flags=cv2.INPAINT_TELEA)

    return imwarped
```

```python
#question 5
import cv2
import numpy as np
import matplotlib.pyplot as plt

def computeH(im1_pts, im2_pts):
    """
    Compute the homography matrix H using corresponding points from two images.

    Parameters:
        im1_pts (numpy.ndarray): n x 2 array of (x, y) points from image 1.
        im2_pts (numpy.ndarray): n x 2 array of (x, y) corresponding points from image 2.

    Returns:
        H (numpy.ndarray): 3x3 homography matrix.
    """
    n = im1_pts.shape[0]
    A = []

    for i in range(n):
        x1, y1 = im1_pts[i]
        x2, y2 = im2_pts[i]

        A.append([-x1, -y1, -1, 0, 0, 0, x2*x1, x2*y1, x2])
        A.append([0, 0, 0, -x1, -y1, -1, y2*x1, y2*y1, y2])

    A = np.array(A)

    # Solve for H using SVD
    U, S, Vt = np.linalg.svd(A)
    H = Vt[-1].reshape(3, 3)

    # Normalize so that h22 = 1
    H = H / H[2, 2]

    return H

# Load images (Fill in your image filenames)
image_path_1 = "/mnt/data/img2warp.jpg"
image_path_2 = "/mnt/data/dog.png"
img1 = cv2.imread(image_path_1)
img2 = cv2.imread(image_path_2)

# Define corresponding points manually (replace these with your actual points)
im1_pts = np.array([[100, 200], [150, 250], [300, 400], [500, 600]])  # Points in image1
im2_pts = np.array([[120, 210], [160, 260], [320, 410], [520, 610]])  # Corresponding points in image2

# Compute homography
H = computeH(im1_pts, im2_pts)
print("Computed Homography Matrix:\n", H)

# Warp image1 using the computed homography
warped_img = cv2.warpPerspective(img1, H, (img2.shape[1], img2.shape[0]))

# Display images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title("Original Image 1")

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(warped_img, cv2.COLOR_BGR2RGB))
plt.title("Warped Image 1 onto Image 2")

plt.show()
```
